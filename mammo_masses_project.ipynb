# Predict whether a mammogram mass is benign or malignant

## Preparing the data

```python
from google.colab import drive
drive.mount('/content/gdrive/')
data_dir = 'gdrive/My Drive/dataset/'
```

```python
import pandas as pd

masses_data = pd.read_csv(data_dir + 'mammographic_masses.data.txt')
masses_data.head()
```

```python
masses_data = pd.read_csv(data_dir + 'mammographic_masses.data.txt', na_values=['?'], names = ['BI-RADS', 'age', 'shape', 'margin', 'density', 'severity'])
masses_data.head()
```

```python
masses_data.describe()
```

```python
masses_data.loc[(masses_data['age'].isnull()) |
              (masses_data['shape'].isnull()) |
              (masses_data['margin'].isnull()) |
              (masses_data['density'].isnull())]
```

The missing data seems randomly distributed, so dropping rows with missing data.

```python
masses_data.dropna(inplace=True)
masses_data.describe()
```

Creating an array that extracts only the feature data we want to work with (age, shape, margin, and density) and another array that contains the classes (severity). Also need an array of the feature name labels.

```python
all_features = masses_data[['age', 'shape',
                             'margin', 'density']].values


all_classes = masses_data['severity'].values

feature_names = ['age', 'shape', 'margin', 'density']

all_features
```

##Normalize the attribute data.

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler()
all_features_scaled = scaler.fit_transform(all_features)
all_features_scaled
```

## Decision Trees

```python
import numpy
from sklearn.model_selection import train_test_split

numpy.random.seed(1234)

(training_inputs,
 testing_inputs,
 training_classes,
 testing_classes) = train_test_split(all_features_scaled, all_classes, train_size=0.75, random_state=1)
```

DecisionTreeClassifier and fitting it to the training data.

```python
from sklearn.tree import DecisionTreeClassifier

clf= DecisionTreeClassifier(random_state=1)

# Train the classifier on the training set
clf.fit(training_inputs, training_classes)
```

Display the resulting decision tree.

```python
from IPython.display import Image  
from sklearn.externals.six import StringIO  
from sklearn import tree
from pydotplus import graph_from_dot_data 

dot_data = StringIO()  
tree.export_graphviz(clf, out_file=dot_data,  
                         feature_names=feature_names)  
graph = graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())  
```

Measure the accuracy of the resulting decision tree model using test data.

```python
clf.score(testing_inputs, testing_classes)
```

Using K-Fold cross validation to get a better measure of your model's accuracy (K=10).

```python
from sklearn.model_selection import cross_val_score

clf = DecisionTreeClassifier(random_state=1)

cv_scores = cross_val_score(clf, all_features_scaled, all_classes, cv=10)

cv_scores.mean()
```

##RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=10, random_state=1)
cv_scores = cross_val_score(clf, all_features_scaled, all_classes, cv=10)

cv_scores.mean()
```

## SVM

Next trying svm.SVC with a linear kernel.

```python
from sklearn import svm

C = 1.0
svc = svm.SVC(kernel='linear', C=C)
```

```python
cv_scores = cross_val_score(svc, all_features_scaled, all_classes, cv=10)

cv_scores.mean()
```

## KNN

```python
from sklearn import neighbors

clf = neighbors.KNeighborsClassifier(n_neighbors=10)
cv_scores = cross_val_score(clf, all_features_scaled, all_classes, cv=10)

cv_scores.mean()
```

Writing a for loop to run KNN with K values ranging from 1 to 50 and see if K makes a substantial difference.

```python
for n in range(1, 50):
    clf = neighbors.KNeighborsClassifier(n_neighbors=n)
    cv_scores = cross_val_score(clf, all_features_scaled, all_classes, cv=10)
    print (n, cv_scores.mean())
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

scaler = preprocessing.MinMaxScaler()
all_features_minmax = scaler.fit_transform(all_features)

clf = MultinomialNB()
cv_scores = cross_val_score(clf, all_features_minmax, all_classes, cv=10)

cv_scores.mean()
```

## Revisiting SVM

svm.SVC may perform differently with different kernels. 

```python
C = 1.0
svc = svm.SVC(kernel='rbf', C=C, gamma = 'scale')
cv_scores = cross_val_score(svc, all_features_scaled, all_classes, cv=10)
cv_scores.mean()
```

```python
C = 1.0
svc = svm.SVC(kernel='sigmoid', C=C, gamma = 'scale')
cv_scores = cross_val_score(svc, all_features_scaled, all_classes, cv=10)
cv_scores.mean()
```

```python
C = 1.0
svc = svm.SVC(kernel='poly', C=C, gamma = 'scale')
cv_scores = cross_val_score(svc, all_features_scaled, all_classes, cv=10)
cv_scores.mean()
```

## Logistic Regression

Fundamentally this is just a binary classification problem. Trying Logisitic Regression is a very simple way to tackle this sort of thing.

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(solver = 'lbfgs')#Algorithm to use in the optimization problem For multiclass problems, only ‘newton-cg’, ‘sag’, ‘saga’ and ‘lbfgs’ handle multinomial loss
cv_scores = cross_val_score(clf, all_features_scaled, all_classes, cv=10)
cv_scores.mean()
```

## Neural Networks

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

def create_model():
    model = Sequential()
    #4 feature inputs going into an 6-unit layer
    model.add(Dense(6, input_dim=4, kernel_initializer='normal', activation='relu'))
    # "Deep learning" turns out to be unnecessary - this additional hidden layer doesn't help either.
    #model.add(Dense(4, kernel_initializer='normal', activation='relu'))
    # Output layer with a binary classification (benign or malignant)
    model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
    # Compile model; adam seemed to work best
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

# Wrap our Keras model in an estimator compatible with scikit_learn
estimator = KerasClassifier(build_fn=create_model, epochs=1000, verbose=0)
# Now we can use scikit_learn's cross_val_score to evaluate this model identically to the others
cv_scores = cross_val_score(estimator, all_features_scaled, all_classes, cv=10)
cv_scores.mean()
```

### The only clear loser is decision trees! Every other algorithm could be tuned to produce comparable results with 79-80% accuracy.
